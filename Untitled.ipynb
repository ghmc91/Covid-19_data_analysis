```python
#Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dadosCoronaBarroso = pd.read_csv('/home/gustavo/Downloads/covid-19-barroso.csv');
dadosCoronaBarroso.tail()
```

```python
dadosCoronaBarbacena = pd.read_csv('/home/gustavo/Downloads/covid-19-barbacena.csv');
dadosCoronaBarbacena.tail()
```

```python
dfBso = dadosCoronaBarroso[['epidemiological_week', 'date','last_available_confirmed']]
dfBso = dfBso.groupby(['epidemiological_week']).max().reset_index()
dfBso['date'] = pd.to_datetime(dfBso.date)
dfBso['date'] = dfBso['date'].dt.strftime('%d/%m')
dfBso = dfBso.rename(columns={'last_available_confirmed':'last_available_confirmed_bso'})
dfBso = pd.DataFrame(dfBso)
dfBso.tail()
```

```python
dfBq = dadosCoronaBarbacena[['epidemiological_week','date','last_available_confirmed']]
dfBq = dfBq.groupby(['epidemiological_week']).max().reset_index()
dfBq['date'] = pd.to_datetime(dfBq.date)
dfBq['date'] = dfBq['date'].dt.strftime('%d/%m')
dfBq = dfBq.rename(columns={'last_available_confirmed':'last_available_confirmed_bq'})
dfBq = pd.DataFrame(dfBq)
dfBq.tail()
```

```python
df_cidades = pd.merge(dfBq, dfBso[['date','last_available_confirmed_bso']], on='date', how='left')
df_cidades
```

```python
fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = (12,6)

ax.plot(df_cidades.date, df_cidades.last_available_confirmed_bso, 
        color='green', marker='o', linestyle='dashed', linewidth=2, 
        markersize=8)


ax.plot(df_cidades.date, df_cidades.last_available_confirmed_bq, 
        color='red', marker='o', linestyle='dashed', linewidth=2, 
        markersize=8)

plt.legend(['Barroso','Barbacena'])

ax.set(ylabel='nmero de casos', xlabel='', title='Evoluo do covid em Barroso-MG/Barbacena \npor semana epidemiolgica')

ax.grid()
fig.savefig('Covid.png')
plt.show()
```
